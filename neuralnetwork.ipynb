```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Model
from keras.layers import Input, Dense
from keras.optimizers import Adam
from keras.callbacks import LearningRateScheduler
import numpy as np

# Load the CSV file
data = pd.read_csv('./banjir_filtered.txt', sep='\t')

# Split data into input features (X) and target label (y)
X = data[['tinggi_air', 'kode_pintu', 'kode_kota']].values
y = data['banjir'].values

# Normalize input features
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# Define the model architecture
input_layer = Input(shape=(3,))
hidden_layer_1 = Dense(32, activation='relu')(input_layer)
hidden_layer_2 = Dense(16, activation='relu')(hidden_layer_1)
output_layer = Dense(1, activation='sigmoid')(hidden_layer_2)

model = Model(inputs=input_layer, outputs=output_layer)

# Define a learning rate schedule
def lr_schedule(epoch):
    learning_rate = 0.001
    if epoch > 10:
        learning_rate *= 0.1
    return learning_rate

# Compile the model with Adam optimizer and learning rate schedule
optimizer = Adam(learning_rate=0.001)
model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])

# Create a learning rate scheduler callback
lr_scheduler = LearningRateScheduler(lr_schedule)

# Train the model with the learning rate scheduler
model.fit(X, y, epochs=100, batch_size=2048, callbacks=[lr_scheduler])

# predict test data

banjir_test  = pd.read_csv("./banjir_test.txt", sep="\t")
banjir_test = banjir_test.dropna()
X_test = banjir_test.drop(["banjir"], axis=1)
y_test = banjir_test["banjir"]

loss, accuracy = model.evaluate(X_test, y_test)
print('Test Loss:', loss)
print('Test Accuracy:', accuracy)

# Make predictions
# probabilities = model.predict(X_test)
# predictions = np.where(probabilities > 0.5, 1, 0)

predictions = model.predict(X_test)

# save prediction
banjir_test["predictions"] = predictions
banjir_test.to_csv("./banjir_test_prediksi_filtered_neural.txt", sep="\t", index=False)
```
